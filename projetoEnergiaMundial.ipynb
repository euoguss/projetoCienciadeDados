
# üìà Previs√£o de Consumo de Energia Mundial por Fonte (50 anos)

Este notebook apresenta a previs√£o de consumo de energia separada por tipo (F√≥ssil, Renov√°vel e Nuclear), utilizando regress√£o linear.

A base de dados cont√©m o consumo de diversas fontes de energia entre os anos de 1983 e 2014. O objetivo √© prever a tend√™ncia at√© 2064.


```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv("data/substituicaoConsumoDeEnergiaMundial.csv")

df["Energia_Fossil"] = df["Coal"] + df["Oil"] + df["Gas"]
df["Energia_Renovavel"] = (
    df["Hydropower"]
    + df["Wind"]
    + df["Solar"]
    + df["Biofuels"]
    + df["Other_renewables"]
    + df["Traditional_biomass"]
)
df["Energia_Nuclear"] = df["Nuclear"]
```

```python

X = df[["Year"]]
y_fossil = df["Energia_Fossil"]
y_renovavel = df["Energia_Renovavel"]
y_nuclear = df["Energia_Nuclear"]

# Treinar modelos
model_fossil = LinearRegression().fit(X, y_fossil)
model_renovavel = LinearRegression().fit(X, y_renovavel)
model_nuclear = LinearRegression().fit(X, y_nuclear)
```

```python

anos_futuros = pd.DataFrame({"Year": list(range(df["Year"].min(), 2073))})
fossil_pred = model_fossil.predict(anos_futuros)
renovavel_pred = model_renovavel.predict(anos_futuros)
nuclear_pred = model_nuclear.predict(anos_futuros)
```

```python

plt.figure(figsize=(10, 5))
plt.plot(df["Year"], y_fossil, label="F√≥ssil - Real", marker='o')
plt.plot(anos_futuros["Year"], fossil_pred, label="F√≥ssil - Previsto", linestyle='--')
plt.title("Consumo de Energia F√≥ssil - Hist√≥rico e Previs√£o (50 anos)")
plt.xlabel("Ano")
plt.ylabel("Consumo de Energia")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 5))
plt.plot(df["Year"], y_renovavel, label="Renov√°vel - Real", marker='o')
plt.plot(anos_futuros["Year"], renovavel_pred, label="Renov√°vel - Previsto", linestyle='--')
plt.title("Consumo de Energia Renov√°vel - Hist√≥rico e Previs√£o (50 anos)")
plt.xlabel("Ano")
plt.ylabel("Consumo de Energia")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 5))
plt.plot(df["Year"], y_nuclear, label="Nuclear - Real", marker='o')
plt.plot(anos_futuros["Year"], nuclear_pred, label="Nuclear - Previsto", linestyle='--')
plt.title("Consumo de Energia Nuclear - Hist√≥rico e Previs√£o (50 anos)")
plt.xlabel("Ano")
plt.ylabel("Consumo de Energia")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df["Year"], y_fossil, label="F√≥ssil - Real", marker='o')
plt.plot(anos_futuros["Year"], fossil_pred, label="F√≥ssil - Previsto", linestyle='--')

plt.plot(df["Year"], y_renovavel, label="Renov√°vel - Real", marker='o')
plt.plot(anos_futuros["Year"], renovavel_pred, label="Renov√°vel - Previsto", linestyle='--')

plt.plot(df["Year"], y_nuclear, label="Nuclear - Real", marker='o')
plt.plot(anos_futuros["Year"], nuclear_pred, label="Nuclear - Previsto", linestyle='--')

plt.title("Previs√£o de Consumo de Energia por Tipo - 50 anos")
plt.xlabel("Ano")
plt.ylabel("Consumo de Energia (unidade do dataset)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
